```python
def dataframe_to_excel(df):
    wb = openpyxl.Workbook()
    ws = wb.active

    for r in dataframe_to_rows(df, index=True, header=True):
        ws.append(r)

    return wb

def excel_to_word(excel_file, word_file):
    # Open the Excel workbook
    workbook = openpyxl.load_workbook(excel_file)
    sheet = workbook.active

    # Create a new Word document
    doc = Document()

    # Get the dimensions of the Excel table
    num_rows = sheet.max_row
    num_cols = sheet.max_column
```

```python
# Define a custom style with bold and italic attributes
def add_custom_style(doc, style_id, name, base_style=None):
    element = OxmlElement('w:style')
    element.set(nsdecls('w'), 'type', 'paragraph')
    element.set(nsdecls('w'), 'styleId', style_id)

    name_element = OxmlElement('w:name')
    name_element.set(nsdecls('w'), 'val', name)
    element.append(name_element)

    if base_style is not None:
        base_style_element = OxmlElement('w:basedOn')
        base_style_element.set(nsdecls('w'), 'val', base_style)
        element.append(base_style_element)

    run_properties = OxmlElement('w:rPr')
    bold_element = OxmlElement('w:b')
    bold_element.set(nsdecls('w'), 'val')
    italic_element = OxmlElement('w:i')
    italic_element.set(nsdecls('w'), 'val')
    run_properties.append(bold_element)
    run_properties.append(italic_element)
    element.append(run_properties)

    doc.styles._element.append(element)
    doc.styles._update_styles()
```
