```python
def dataframe_to_excel(df):
    wb = openpyxl.Workbook()
    ws = wb.active

    for r in dataframe_to_rows(df, index=True, header=True):
        ws.append(r)

    return wb

def excel_to_word(excel_file, word_file):
    # Open the Excel workbook
    workbook = openpyxl.load_workbook(excel_file)
    sheet = workbook.active

    # Create a new Word document
    doc = Document()

    # Get the dimensions of the Excel table
    num_rows = sheet.max_row
    num_cols = sheet.max_column
```

```python
import pandas as pd

import openpyxl

def compare_and_save_differences(sheet_name, df1, df2, writer):
    diff_df = pd.concat([df1, df2]).drop_duplicates(keep=False)
    if not diff_df.empty:
        diff_df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Differences for sheet '{sheet_name}' saved in the Excel file")
    else:
        print(f"No differences found for sheet '{sheet_name}'")

# Load the Excel files into DataFrames

file1 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture- SAM-PAM Risk Assessment (7).xlsx'

file2 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture  SAM-PAM Risk Assessment (v.2 Pre-PPQ)v.2.xlsx'

output_file = 'differences_combined_2.xlsx'

wb1 = openpyxl.load_workbook(file1, data_only=True)
wb2 = openpyxl.load_workbook(file2, data_only=True)


# Function to find and print cell changes
def find_cell_changes(sheet_name, sheet1, sheet2):
    for row in range(1, sheet1.max_row + 1):
        for col in range(1, sheet1.max_column + 1):
            cell1 = sheet1.cell(row=row, column=col)
            cell2 = sheet2.cell(row=row, column=col)
            if cell1.value != cell2.value:
                cell_address = cell1.coordinate
                print(f"Change detected in sheet '{sheet_name}', cell {cell_address}")

# Compare sheets and find cell changes
for sheet_name in wb1.sheetnames:
    if sheet_name in wb2.sheetnames:
        sheet1 = wb1[sheet_name]
        sheet2 = wb2[sheet_name]
        find_cell_changes(sheet_name, sheet1, sheet2)
    else:
        print(f"Sheet '{sheet_name}' not found in {file2}")

wb1.close()
wb2.close()
```

```python
file1 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture- SAM-PAM Risk Assessment (7).xlsx'
file2 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture  SAM-PAM Risk Assessment (v.2 Pre-PPQ)v.2.xlsx'

df1 = pd.read_excel(file1, sheet_name=None)
df2 = pd.read_excel(file2, sheet_name=None)

output_file = r'M:\CC Process Development\AntiIL5EP\Control Strategy\differences_combined_2.xlsx'

def compare_difference(df1, df2):
    diff_dict = {}
    for column in df1.columns:
        if column in df2.columns:
            compared = df1[column].reset_index(drop=True).compare(df2[column].reset_index(drop=True),result_names=("File_1","File_2"))
            if len(compared) != 0:
                diff_dict[column] = compared
        else:
            print(f"Column: {column}, not found in second file")
    df_final = pd.concat(diff_dict,ignore_index=False)
    return df_final


def data_check_excel(excel_1,excel_2):
    final_dict = {}
    for key, data in excel_1.items():
        if key in excel_2.keys():
            df = compare_difference(excel_1[key], excel_2[key])
            final_dict[key] = df

    writer=pd.ExcelWriter(output_file) 
    for key in final_dict:
        final_dict[key].to_excel(writer,sheet_name=key)
    return final_dict

data_check_excel(df1,df2)
```
