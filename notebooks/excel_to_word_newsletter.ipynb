```python
import pandas as pd
import numpy as np

import openpyxl
from openpyxl.styles import Font
from openpyxl.utils.dataframe import dataframe_to_rows

from docx import Document
from docx.shared import RGBColor
from docx.shared import Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
```

```python
df = pd.read_excel(r"M:\Zach Hatzenbeller\newsletter_scripts\Convert Excel to Word_v6_speediest.xlsm")
df["Asset"] = pd.Series(df["Asset"]).fillna(method='ffill')
df.head()
```

```python
def dataframe_to_excel(df):
    wb = openpyxl.Workbook()
    ws = wb.active

    for r in dataframe_to_rows(df, index=True, header=True):
        ws.append(r)

    return wb

def excel_to_word(excel_file, word_file):
    # Open the Excel workbook
    workbook = openpyxl.load_workbook(excel_file)
    sheet = workbook.active

    # Create a new Word document
    doc = Document()

    # Get the dimensions of the Excel table
    num_rows = sheet.max_row
    num_cols = sheet.max_column
```

```python
worksheet = dataframe_to_excel(df)

# Open the Excel workbook
sheet = worksheet.active

# Create a new Word document
doc = Document()

# Get the dimensions of the Excel table
num_rows = sheet.max_row
num_cols = sheet.max_column

len(df["Asset"].unique())

# Create a Word table with the same dimensions as the Excel table
table = doc.add_table(rows=num_rows, cols=2)
table.style = 'Table Grid'
table.allow_autofit = False

yellow = openpyxl.styles.colors.Color(rgb='FFFF00')

# Loop through each cell in the Excel table and copy its value and format to the Word table
table_row = 3
for count, row in enumerate(sheet.iter_rows(min_row=3, max_row=num_rows-1, min_col=1, max_col=num_cols)):
    
    paragraph_asset = table.cell(row_idx=count+3, col_idx=0).add_paragraph()
    run_asset = paragraph_asset.add_run(str(sheet.cell(row=table_row, column=2).value))

    # Apply font styling to the run
    font = run_asset.font
    font.name = 'Arial Narrow'
    # Set the font size (e.g., 12pt)
    font.size = Pt(12)
    # Set the font color (e.g., red)
    font.color.rgb = RGBColor(0, 0, 0)  # RGBColor(red, green, blue)
    # Apply bold and italic styles
    font.bold = True
    font.italic = False
    font.underline = False
    
    paragraph_asset.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

    paragraph = table.cell(row_idx=count+3, col_idx=1).add_paragraph()

    # Add text to the paragraph
    run_head1 = paragraph.add_run(str(sheet.cell(row=1, column=4).value))
    paragraph.add_run("\n")
    run_up1 = paragraph.add_run("Upstream: ")
    run_text1 = paragraph.add_run(str(sheet.cell(table_row,4).value))
    paragraph.add_run("\n")

    run_down1 = paragraph.add_run("Downstream: ")
    run_text2 = paragraph.add_run(str(sheet.cell(table_row+1,4).value))
    paragraph.add_run("\n")
    
    run_vc1 = paragraph.add_run("Viral Clearance: ")
    run_text3 = paragraph.add_run(str(sheet.cell(table_row+2,4).value))
    paragraph.add_run("\n")


    run_head2 = paragraph.add_run(str(sheet.cell(row=1, column=5).value))
    paragraph.add_run("\n")
    run_up2 = paragraph.add_run("Upstream: ")
    run_text4 = paragraph.add_run(str(sheet.cell(table_row,5).value))
    paragraph.add_run("\n")

    run_down2 = paragraph.add_run("Downstream: ")
    run_text5 = paragraph.add_run(str(sheet.cell(table_row+1,5).value))
    paragraph.add_run("\n")
    
    run_vc2 = paragraph.add_run("Viral Clearance: ")
    run_text6 = paragraph.add_run(str(sheet.cell(table_row+2,5).value))
    paragraph.add_run("\n")


    run_head3 = paragraph.add_run(str(sheet.cell(row=1, column=6).value))
    paragraph.add_run("\n")
    run_up3 = paragraph.add_run("Upstream: ")
    run_text7 = paragraph.add_run(str(sheet.cell(table_row,6).value))
    paragraph.add_run("\n")

    run_down3 = paragraph.add_run("Downstream: ")
    run_text8 = paragraph.add_run(str(sheet.cell(table_row+1,6).value))
    paragraph.add_run("\n")
    
    run_vc3 = paragraph.add_run("Viral Clearance: ")
    run_text9 = paragraph.add_run(str(sheet.cell(table_row+2,6).value))
    paragraph.add_run("\n")


    run_head4 = paragraph.add_run(str(sheet.cell(row=1, column=7).value))
    paragraph.add_run("\n")
    run_up4 = paragraph.add_run("Upstream: ")
    run_text10 = paragraph.add_run(str(sheet.cell(table_row,7).value))
    paragraph.add_run("\n")

    run_down4 = paragraph.add_run("Downstream: ")
    run_text11 = paragraph.add_run(str(sheet.cell(table_row+1,7).value))
    paragraph.add_run("\n")
    
    run_vc4 = paragraph.add_run("Viral Clearance: ")
    run_text12 = paragraph.add_run(str(sheet.cell(table_row+2,7).value))
    paragraph.add_run("\n")

    table_row += 3

    run_heads = [run_head1,run_head2,run_head3,run_head4]
    run_up = [run_up1,run_up2,run_up3,run_up4]
    run_down = [run_down1,run_down2,run_down3,run_down4]
    run_vc = [run_vc1,run_vc2,run_vc3,run_vc4]
    run_text = [run_text1,run_text2,run_text3,run_text4,run_text5,run_text6,run_text7,run_text8,run_text9,run_text10,run_text11,run_text12]


    for i in run_heads:
        # Apply font styling to the run
        font = i.font
        font.name = 'Arial Narrow'
        # Set the font size (e.g., 12pt)
        font.size = Pt(12)
        # Set the font color (e.g., red)
        font.color.rgb = RGBColor(0, 0, 0)  # RGBColor(red, green, blue)
        # Apply bold and italic styles
        font.bold = True
        font.italic = False
        font.underline = True
    
    for i in run_up:
        # Apply font styling to the run
        font = i.font
        font.name = 'Arial Narrow'
        # Set the font size (e.g., 12pt)
        font.size = Pt(10)
        # Set the font color (e.g., red)
        font.color.rgb = RGBColor(255, 102, 0)  # RGBColor(red, green, blue)
        # Apply bold and italic styles
        font.bold = True
        font.italic = True

    for i in run_down:
        # Apply font styling to the run
        font = i.font
        font.name = 'Arial Narrow'
        # Set the font size (e.g., 12pt)
        font.size = Pt(10)
        # Set the font color (e.g., red)
        font.color.rgb = RGBColor(3, 37, 126)  # RGBColor(red, green, blue)
        # Apply bold and italic styles
        font.bold = True
        font.italic = True
    
    for i in run_vc:
        # Apply font styling to the run
        font = i.font
        font.name = 'Arial Narrow'
        # Set the font size (e.g., 12pt)
        font.size = Pt(10)
        # Set the font color (e.g., red)
        font.color.rgb = RGBColor(128, 0, 128)  # RGBColor(red, green, blue)
        # Apply bold and italic styles
        font.bold = True
        font.italic = True
    
    for i in run_text:
        # Apply font styling to the run
        font = i.font
        font.name = 'Arial Narrow'
        # Set the font size (e.g., 12pt)
        font.size = Pt(10)
        # Set the font color (e.g., red)
        font.color.rgb = RGBColor(0, 0, 0)  # RGBColor(red, green, blue)
        
doc.save(r"M:\Zach Hatzenbeller\newsletter_scripts\Test.docx")
```

```python
import pandas as pd

# Load the Excel files into DataFrames
file1 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture- SAM-PAM Risk Assessment (7).xlsx'
file2 = r'M:\CC Process Development\AntiIL5EP\Control Strategy\GSK3511294 (depemokimab) Drug Substance Manufacture  SAM-PAM Risk Assessment (v.2 Pre-PPQ)v.2.xlsx'

xl1 = pd.ExcelFile(file1)
xl2 = pd.ExcelFile(file2)

df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

output_file = r'M:\CC Process Development\AntiIL5EP\Control Strategy\differences.xlsx'
writer = pd.ExcelWriter(output_file)

for sheet in xl1.sheet_names:
    df1 = pd.read_excel(file1, sheet_name=sheet)
    if sheet in xl2.sheet_names:
        df2 = pd.read_excel(file2, sheet_name=sheet)

        # Compare the two DataFrames and find differences
        diff_df = pd.concat([df1, df2]).drop_duplicates(keep=False)

        # Save the differences to a new Excel file
        diff_df.to_excel(writer, index=False, sheet_name=sheet)

print(f"Differences saved to {output_file}")
```
